# **Student Result Analyzer
**
Problem Statement
You are given the marks of a student in 5 subjects. Write a Python program that:

Stores the student's name (string) and roll number (integer) in variables.
Takes 5 subject marks as a list of floats.
Calculates the total and average marks.
Assigns a grade using the following rules:
    **average >= 90 → Grade A
    average >= 75 → Grade B
    average >= 60 → Grade C
    average >= 40 → Grade D
    else → Fail**
Also prints which subjects the student scored below 40 in (use a loop).
Wrap all logic inside a function analyze_result(name, roll, marks).

```python
def analyze_result(name, roll, marks):

    total = sum(marks)
    average = total / len(marks)

    if average >= 90:
        grade = "A"
    elif average >= 75:
        grade = "B"
    elif average >= 60:
        grade = "C"
    elif average >= 40:
        grade = "D"
    else:
        grade = "Fail"

    print("Student Result")
    print(f"Student Name : {name}")
    print(f"Roll Number  : {roll}")
    print(f"Marks        : {marks}")
    print(f"Total Marks  : {total}")
    print(f"Average      : {average:.2f}")
    print(f"Grade        : {grade}")

    print("\nSubjects Below 40")

    found = False

    for i in range(len(marks)):
        if marks[i] < 40:
            print(f"Subject {i+1} : {marks[i]}")
            found = True

    if not found:
        print("None")


name = input("Enter Student Name : ")
roll = int(input("Enter Roll Number : "))

marks = []

print("\nEnter Marks of 5 Subjects")

for i in range(5):
    mark = float(input(f"Subject {i+1}: "))
    marks.append(mark)

analyze_result(name, roll, marks)
```
