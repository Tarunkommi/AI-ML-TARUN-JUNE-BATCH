Variables

A variable is a named location used to store data in memory.

Rules
.Must start with a letter or underscore.
.Cannot start with a number.
.Case-sensitive.
.No spaces allowed.

```python
student_name = "Tarun"
roll_no = 101
print("Student Name:", student_name)
print("Roll Number:", roll_no)
```

 Data Types

Python automatically detects the type of data.

Data Type	       Example	     Mutable
int            	     10	           ❌
float	             95.5	       ❌
str	               "Hello"	       ❌
bool	            True	       ❌

```python
age = 20
marks = 88.5
name = "Tarun"
print(f"name={name} is {age} years old and scored {marks} marks.")
```

List

A list stores multiple values in order.

Properties

.Ordered
.Mutable
.Allows duplicates


List Methods
Method	                    Description	                          Example
append(x)	          Adds one element at the end	            numbers.append(40)
extend(iterable)	  Adds multiple elements	                numbers.extend([50,60])
insert(index, item)	  Inserts at specified index	            numbers.insert(1,15)
remove(item)	      Removes first occurrence	                numbers.remove(20)
pop()	              Removes last element	                    numbers.pop()
pop(index)	          Removes element at index	                numbers.pop(1)
clear()	              Removes all elements	                    numbers.clear()
index(item)	          Returns index of item	                    numbers.index(30)
count(item)	          Counts occurrences	                    numbers.count(20)
sort()	              Sorts ascending	                        numbers.sort()
sort(reverse=True)	  Sorts descending	                        numbers.sort(reverse=True)
reverse()	          Reverses list	                            numbers.reverse()
copy()	              Returns shallow copy	                    new = numbers.copy()


Built-in Functions for Lists
Function	  Description
len()	    Number of elements
max()	    Largest element
min()	    Smallest element
sum()	    Sum of elements
sorted()	Returns sorted list
any()	    True if any element is True
all()	    True if all elements are True
list()	    Converts iterable to list

```python
marks = [88, 92, 76, 95, 60]
print(marks[0])
marks.append(100)
print(marks)
marks[2] = 80
print(marks)
```

Tuple

A tuple is similar to a list but cannot be modified.

Properties

.Ordered
.Immutable
.Faster than list

Tuple Methods

Only 2 methods exist.

Method	  Description        Example
count()	Counts occurrences	t.count(10)
index()	Finds first index	t.index(20)

Built-in Functions for Tuples
Function	Description
len()	     Length
max()	     Largest value
min()	     Smallest value
sum()	     Sum
sorted()	 Returns sorted list
tuple()	     Convert iterable to tuple
any()	     Any True
all()	     All True

```python
book = ("Python", "Guido", 1991)
print(book)
book[0] = "Java"
print(book)
```

Dictionary

Stores data as key-value pairs

Properties

.Mutable
.Fast lookup
.Keys must be unique

Dictionary Methods
Method	       Description	                      Example
clear()	     Removes all items	                d.clear()
copy()	     Creates shallow copy	            d.copy()
fromkeys() 	 Creates dictionary with keys	    dict.fromkeys(["a","b"],0)
get()	     Returns value safely	            d.get("name")
items()	     Returns key-value pairs	        d.items()
keys()	     Returns keys	                    d.keys()
values()	 Returns values	                    d.values()
pop(key)	 Removes key	                    d.pop("age")
popitem()	 Removes last inserted item	        d.popitem()
setdefault() Returns value or inserts key	    d.setdefault("city","Hyd")
update()	 Updates dictionary	                d.update({"age":21})

Built-in Functions
Function	Description
len()	  Number of key-value pairs
dict()	  Creates dictionary
sorted()  Sorts keys
any()	  True if any key is True
all()	  True if all keys are True

```python
student = {
    "name":"Tarun",
    "roll":101
}
print(student["name"])
student["cgpa"] = 8.0
print(student)
```

Set

Stores unique values.

Properties

Unordered
Mutable
Unique values only

Set Methods
Method	                     Description	                    Example
add()	                Adds one element	                    s.add(50)
update()	            Adds multiple elements	                s.update([60,70])
remove()	            Removes element	                        s.remove(20)
discard()	            Removes without error	                s.discard(20)
pop()	                Removes random element	                s.pop()
clear()	                Removes everything	                    s.clear()
copy()	                Creates shallow copy	                s.copy()
union()         	    Combines sets	                        A.union(B)
intersection()  	    Common elements	                        A.intersection(B)
difference()	        Elements in first set only	            A.difference(B)
symmetric_difference()	Elements in either set, not both	    A.symmetric_difference(B)
issubset()	            Checks subset	                        A.issubset(B)
issuperset()	        Checks superset	                        A.issuperset(B)
isdisjoint()	        No common elements	                    A.isdisjoint(B)
intersection_update()	Updates with intersection	            A.intersection_update(B)
difference_update()	    Updates with difference	                A.difference_update(B)
symmetric_difference_update()	Updates with symmetric difference	A.symmetric_difference_update(B)


Built-in Functions
Function	    Description
len()	    Number of elements
max()	    Largest element
min()	    Smallest element
sum()	    Sum of numeric elements
sorted()	Returns sorted list
set()	    Creates a set
any()	    Any element is True
all()	    All elements are True

```python
members = {101,102,103}
members.add(104)
members.add(101)
# Nothing happens because duplicates are not allowed.
print(members)
```
