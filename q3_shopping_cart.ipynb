# Shopping Cart with Default & Mutable Pitfall
### Problem Statement
> This question tests the trickiest topic we covered — mutable default arguments & pass-by-reference behavior.

Part A — Spot the Bug
The following function has a hidden bug. Predict the output and explain why:

def add_item(item, cart=[]):
    cart.append(item)
    return cart

print(add_item("apple"))
print(add_item("banana"))
print(add_item("milk", cart=["bread"]))
print(add_item("eggs"))
Part B — Fix It
Rewrite add_item the correct way so that calling it without a cart starts with a fresh empty list every time.
    
Part C — Build the Cart
Now build a complete shopping cart program:

1.Function create_cart(owner, discount=0) → returns a dict like {"owner": owner, "items": [], "discount": discount}. Here discount is an immutable default — safe.
2.Function add_to_cart(cart, name, price, qty=1) → appends {"name": name, "price": price, "qty": qty} to cart["items"].
3.Function update_price(price_tuple, new_price) → attempt to modify a tuple element. What happens? Explain in a comment why tuples are immutable.
4.Function calculate_total(cart) → loops over items, sums price * qty, applies discount %, returns final total.
5.Demonstrate calling these functions for two different customers and show their independent carts (proves the Part B fix works).

```python
def create_cart(owner, discount=0):

    return {
        "owner": owner,
        "items": [],
        "discount": discount
    }

def add_to_cart(cart, name, price, qty=1):

    item = {
        "name": name,
        "price": price,
        "qty": qty
    }

    cart["items"].append(item)


def update_price(price_tuple, new_price):

    try:
        price_tuple[0] = new_price

    except TypeError:

        print("Tuple is immutable.")
        print("Cannot modify tuple elements.")


def calculate_total(cart):

    total = 0

    for item in cart["items"]:

        total += item["price"] * item["qty"]

    discount = cart["discount"]

    total = total - (total * discount / 100)

    return total

cart1 = create_cart("Tarun",10)

add_to_cart(cart1,"Laptop",50000,1)
add_to_cart(cart1,"Mouse",1000,2)

cart2 = create_cart("Rahul",5)

add_to_cart(cart2,"Phone",30000,1)
add_to_cart(cart2,"Headphones",2500,1)

print(cart1)
print()

print(cart2)
print()

print("Tarun Total :",calculate_total(cart1))
print("Rahul Total :",calculate_total(cart2))

price = (50000,)

update_price(price,45000)
```
