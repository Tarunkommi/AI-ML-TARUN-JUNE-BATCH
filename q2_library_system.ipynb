# Library Book Management System

### Problem Statement
> You are building a small Library System. Implement the following using the right data structure for each requirement:

Store the library catalog as a dictionary where key = book_id (int) and value = tuple (title, author, year). (Why tuple? Because book details should be immutable.)
Maintain a list called borrowed_books to track currently issued book IDs (insertion order matters).
Maintain a set called members to store unique member IDs (no duplicates allowed).
Write the following functions:
> add_book(catalog, book_id, title, author, year)
> 
> borrow_book(catalog, borrowed_books, book_id) — only if book exists & not already borrowed.
> 
> return_book(borrowed_books, book_id)
> 
> register_member(members, member_id) — should silently ignore duplicates.
> 
> show_available(catalog, borrowed_books) — print all books that are NOT in borrowed_books.
> 
> In main(), demonstrate adding 4 books, registering 3 members (try adding the same member twice), borrowing 2 books, returning 1, and finally displaying available books.

```python
catalog = {}

borrowed_books = []
members = set()

def add_book(catalog, book_id, title, author, year):

    catalog[book_id] = (title, author, year)

    print(f"Book '{title}' added successfully.")

def register_member(members, member_id):

    members.add(member_id)

    print(f"Member {member_id} registered.")

def borrow_book(catalog, borrowed_books, book_id):

    if book_id not in catalog:
        print("Book does not exist.")
        return

    if book_id in borrowed_books:
        print("Book already borrowed.")
        return

    borrowed_books.append(book_id)

    print("Book borrowed successfully.")

def return_book(borrowed_books, book_id):

    if book_id in borrowed_books:
        borrowed_books.remove(book_id)
        print("Book returned.")
    else:
        print("Book was not borrowed.")

def show_available(catalog, borrowed_books):

    print("\nAvailable Books\n")

    for book_id, details in catalog.items():

        if book_id not in borrowed_books:

            title, author, year = details

            print(book_id, title, author, year)

def main():

    add_book(catalog,101,"Python Basics","Guido",2022)
    add_book(catalog,102,"Data Structures","Mark",2021)
    add_book(catalog,103,"Machine Learning","Andrew",2020)
    add_book(catalog,104,"Artificial Intelligence","Russell",2019)

    register_member(members,1)
    register_member(members,2)
    register_member(members,3)
    register_member(members,2)

    borrow_book(catalog,borrowed_books,101)
    borrow_book(catalog,borrowed_books,103)

    return_book(borrowed_books,101)

    show_available(catalog,borrowed_books)

main()
```
